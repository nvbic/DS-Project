```python


def five_fold_CV(clf, params, x, y):
    cv_model = RandomizedSearchCV(clf, params, scoring='roc_auc', n_jobs=-1,
                                  cv=StratifiedKFold(n_splits=5, shuffle=True, random_state=2020), 
                                  verbose=1,n_iter=30, refit=True)
    
    cv_model.fit(x, y)
    
    return cv_model.best_estimator_, cv_model.best_score_
```

```python
# Random Forest

rf = RandomForestClassifier(random_state=2020)

parameters = {'max_features':['auto','sqrt','log2', None],
                    'n_estimators':[100,250, 400,750, 1000], 
                    'min_samples_leaf':[1, 50, 100], 
                    'max_leaf_nodes': [None, 5, 10, 16],
                    'oob_score': [True],
                    'criterion':['gini', 'entropy'], 
                    'min_samples_split':[2, 3, 4, 5], 
                    'max_depth': [None, 2, 8, 16, 20, 24],
                    'min_impurity_decrease':[0.0, 0.01, 0.001, 0.0001, 0.00001],
                    'bootstrap': [True, False],
                    'class_weight': [None, 'balanced','balanced_subsample']}

best_rf_model, best_cv_performance = five_fold_CV(rf, parameters, X_train, y_train)
best_rf_model, best_cv_performance
```
